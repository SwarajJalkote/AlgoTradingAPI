```python
import pandas as pd
import numpy as np
import requests
from pprint import pprint
```

```python
class ScriptDataClass():
    
    def __init__(self, base_url, apikey):
        self.base_url = base_url
        self.apikey = apikey
        self.data = ''
        self.df = ''
    
    def fetch_intraday_data(self, scrip):
        params = {
            'function': 'TIME_SERIES_INTRADAY',
            'symbol': scrip,
            'interval': '5min',
            'apikey': self.apikey
        }
        
        response = requests.get(url = self.base_url, params = params)
#         print("response: ", response.status_code, type(response.status_code))

        if response.status_code == 200:
            self.data = response.json()
#             pprint(data[f'Time Series ({params["interval"]})'])
        else:
            print("Error occured while fetching the data")
            print("Error: ", response)
        
    
    
    def convert_intraday_data(self, scrip):
        '''
            i. timestamp (data type: pandas.Timestamp)
            ii. open (data type: float)
            iii. high (data type: float)
            iv. low(data type: float)
            v. close (data type: float)
            vi. volume (data type: int)
        '''
        params = {
            'function': 'TIME_SERIES_INTRADAY',
            'symbol': scrip,
            'interval': '5min',
            'apikey': self.apikey
        }
        time_data = {'timestamp':[], 'open':[], 'high':[], 'low':[], 'close':[], 'volume':[]}
        json_data = self.data[f'Time Series ({params["interval"]})']
#         df = pd.DataFrame(columns=datatypes)
        for _ in json_data:
            time_data['timestamp'].append(_)
            for key, value in json_data[_].items():
                if key.strip('0123456789. ') in time_data:
                    time_data[key.strip('0123456789. ')].append(value)
                
                
        self.df = pd.DataFrame(time_data)
        
   
```

```python
obj = ScriptDataClass('https://www.alphavantage.co/query', 'E6UM8KIJ2CNPDR3L')
```

```python
obj.fetch_intraday_data('IBM')
# 
```

```python
obj.convert_intraday_data('IBM')
```
